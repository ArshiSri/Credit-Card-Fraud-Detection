```python
#dividing itno training and testing dataset
V_train, V_test, T_train, T_test = train_test_split(V, T, test_size=0.2, random_state = 2, stratify = T)
```

```python
print(V.shape, V_train.shape, V_test.shape)
```

```python
print(T.shape, T_train.shape, T_test.shape)
```

```python
#Model Building
lr = LogisticRegression()
```

```python
lr.fit(V_train, T_train)
```

```python
A = lr.predict(V_train)
B = accuracy_score(A, T_train)
```

```python
C = lr.predict(V_test)
D = accuracy_score(C, T_test)
```
